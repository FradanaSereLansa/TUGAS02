```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,20))
plt.plot(iris['Sisa Bulan Lalu'], label='Sisa Bulan Lalu')
plt.plot(iris['Perkara Masuk'], label='Perkara Masuk')
plt.plot(iris['Putus'], label='Putus')
plt.plot(iris['Minutasi'], label='Minutasi')
plt.plot(iris['Belum Minutasi'], label='Belum Minutasi')
plt.plot(iris['Sisa'], label='Sisa')
plt.xlabel('Data Count')
plt.ylabel('Value')
plt.title('Plot of each parameter')
plt.legend()
plt.show()
```
